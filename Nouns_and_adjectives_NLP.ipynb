```python
import pandas as pd
import nltk
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
df = pd.read_csv('data.csv', encoding='unicode_escape', nrows=10000)
# df = pd.read_csv('data.csv', encoding='unicode_escape')

df
```

```python
# Create a new empty column

df['Noun'] = None

df
```

```python
# Create a new dataframe

df_new = pd.DataFrame()
```

```python
stemmer = nltk.stem.snowball.SnowballStemmer("english")
```

```python
for _, row in df.iterrows():

    try:

        # Store the row's product description and change it to lowercase
        sentence = row['Description']
        sentence = sentence.lower()

        # Classify each product description word (gramatically)
        classified_words = nltk.pos_tag(nltk.word_tokenize(sentence))

        print('\nThis is the word classification:')
        print(classified_words)

        # Loop through each word and only save "nouns"
        for word in classified_words:

            if word[1] == 'NN':

                print('This word is a noun and will be appended to the new df: {}'.format(word[0]))
                row['Noun'] = word[0]
                df_new = df_new.append(row, ignore_index=True, sort=True)

    except:
        print("An exception occurred with this row: {}".format(row))
```

```python
df_new
```

```python
# Remove possible duplicates
df_new = df_new.drop_duplicates()
```

```python
# Save new CSV file
df_new.to_csv('output.csv', index=False)
```
